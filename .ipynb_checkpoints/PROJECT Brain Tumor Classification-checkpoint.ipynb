### Load Modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

### Prepare/collect data

```python
import os

path = os.listdir('brain_tumor/Training/')
classes = {'no_tumor':0, 'pituitary_tumor':1}
```

```python
import cv2
X = []
Y = []
for cls in classes:
    pth = 'brain_tumor/Training/'+cls
    for j in os.listdir(pth):
        img = cv2.imread(pth+'/'+j, 0)
        img = cv2.resize(img, (200,200))
        X.append(img)
        Y.append(classes[cls])
```

```python
X = np.array(X)
Y = np.array(Y)

X_updated = X.reshape(len(X), -1)
```

```python
np.unique(Y)
```

```python
pd.Series(Y).value_counts()
```

```python
X.shape, X_updated.shape
```

### Visualize data

```python
plt.imshow(X[0], cmap='gray')
```

### Prepare data

```python
X_updated = X.reshape(len(X), -1)
X_updated.shape
```

### Split Data

```python
xtrain, xtest, ytrain, ytest = train_test_split(X_updated, Y, random_state=10,
                                               test_size=.20)
```

```python
xtrain.shape, xtest.shape
```

### Feature Scaling

```python
print(xtrain.max(), xtrain.min())
print(xtest.max(), xtest.min())
xtrain = xtrain/255
xtest = xtest/255
print(xtrain.max(), xtrain.min())
print(xtest.max(), xtest.min())
```

### Feature Selection: PCA

```python
from sklearn.decomposition import PCA
```

```python
print(xtrain.shape, xtest.shape)

pca = PCA(.98)
# pca_train = pca.fit_transform(xtrain)
# pca_test = pca.transform(xtest)
pca_train = xtrain
pca_test = xtest
```

### Train Model

```python
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings('ignore')

lg = LogisticRegression(C=0.1)
lg.fit(xtrain, ytrain)
```

```python
from sklearn.svm import SVC
sv = SVC()
sv.fit(xtrain, ytrain)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(xtrain, ytrain)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

xtrain_cnn = xtrain.reshape(xtrain.shape[0], 200, 200, 1)
xtest_cnn = xtest.reshape(xtest.shape[0], 200, 200, 1)

xtrain_cnn = xtrain_cnn / 255.0
xtest_cnn = xtest_cnn / 255.0

model = Sequential()
model.add(Conv2D(32, (3, 3), input_shape=(200, 200, 1), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(xtrain_cnn, ytrain, epochs=10, batch_size=32, validation_data=(xtest_cnn, ytest))
```

### Evaluation

```python
print("Training Score:", lg.score(xtrain, ytrain))
print("Testing Score:", lg.score(xtest, ytest))
```

```python
print("Training Score:", sv.score(xtrain, ytrain))
print("Testing Score:", sv.score(xtest, ytest))
```

```python
print("Training Score:", knn.score(xtrain, ytrain))
print("Testing Score:", knn.score(xtest, ytest))
```

```python
print("Training Score:", model.evaluate(xtrain_cnn, ytrain)[1])
print("Testing Score:", model.evaluate(xtest_cnn, ytest)[1])
```

```python
models = ['Logistic Regression', 'SVM', 'KNN', 'CNN']
train_accuracies = [lg.score(xtrain, ytrain), sv.score(xtrain, ytrain), knn.score(xtrain, ytrain), model.evaluate(xtrain_cnn, ytrain)[1]]
test_accuracies = [lg.score(xtest, ytest), sv.score(xtest, ytest), knn.score(xtest, ytest), model.evaluate(xtest_cnn, ytest)[1]]

plt.figure(figsize=(10, 5))
plt.barh(models, train_accuracies, color='skyblue')
plt.xlabel('Accuracy')
plt.title('Training Accuracies of Different Models')
plt.xlim(0, 1)
plt.show()

plt.figure(figsize=(10, 5))
plt.barh(models, test_accuracies, color='lightcoral')
plt.xlabel('Accuracy')
plt.title('Testing Accuracies of Different Models')
plt.xlim(0, 1)
plt.show()
```

```python
from sklearn.metrics import precision_score, recall_score, confusion_matrix, precision_recall_curve
import seaborn as sns

def plot_confusion_matrix(y_true, y_pred, classes, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 6))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=classes, yticklabels=classes)
    plt.title(title)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()

def plot_precision_recall(y_true, y_pred_proba, title):
    precision, recall, _ = precision_recall_curve(y_true, y_pred_proba)
    plt.figure(figsize=(8, 8))
    plt.plot(recall, precision, color='b', linewidth=2)
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title(title)
    plt.show()

y_pred_lr = lg.predict(xtest)
y_pred_proba_lr = lg.predict_proba(xtest)[:, 1]

y_pred_svm = sv.predict(xtest)

y_pred_knn = knn.predict(xtest)

y_pred_cnn_prob = model.predict(xtest_cnn)
y_pred_cnn = (y_pred_cnn_prob > 0.5).astype(int)

if y_pred_cnn_prob.shape[1] > 1:
    y_pred_cnn_prob = y_pred_cnn_prob[:, 1]
```

```python
plt.figure(figsize=(10, 8))
plot_confusion_matrix(ytest, y_pred_lr, classes=['no_tumor', 'pituitary_tumor'], title='Logistic Regression Confusion Matrix')
```

```python
plt.figure(figsize=(10, 8))
plot_confusion_matrix(ytest, y_pred_svm, classes=['no_tumor', 'pituitary_tumor'], title='SVM Confusion Matrix')
```

```python
plt.figure(figsize=(10, 8))
plot_confusion_matrix(ytest, y_pred_knn, classes=['no_tumor', 'pituitary_tumor'], title='KNN Confusion Matrix')
```

```python
plt.figure(figsize=(10, 8))
plot_confusion_matrix(ytest, y_pred_cnn, classes=['no_tumor', 'pituitary_tumor'], title='CNN Confusion Matrix')
```

```python
plt.figure(figsize=(10, 8))
plot_precision_recall(ytest, y_pred_proba_lr, title='Logistic Regression Precision-Recall Curve')

plt.figure(figsize=(10, 8))
plot_precision_recall(ytest, y_pred_cnn_prob, title='CNN Precision-Recall Curve')
```
